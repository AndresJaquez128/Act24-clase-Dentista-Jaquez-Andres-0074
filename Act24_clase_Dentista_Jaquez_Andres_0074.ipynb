```python



class Dentista:
    def __init__(self, nombre, especialidad, consultorio, experiencia_anos):
        self.nombre = nombre
        self.especialidad = especialidad
        self.consultorio = consultorio
        self.experiencia_anos = experiencia_anos

    def examinar_boca(self):
        return f"🦷 El Dr. {self.nombre} está revisando la dentadura en el consultorio {self.consultorio}."

    def realizar_limpieza(self):
        return f"✨ Iniciando limpieza profunda. Técnica especializada en {self.especialidad}."

    def colocar_ortodoncia(self):
        return f"😬 El Dr. {self.nombre} está ajustando los brackets. ({self.experiencia_anos} años de experiencia garantizados)."

# --- Creación de Objetos ---
dentista1 = Dentista("Roberto Gómez", "Odontología General", "A-101", 12)
dentista2 = Dentista("Elena Torres", "Ortodoncia e Implantología", "B-205", 8)

# --- Sistema Interactivo ---
def ejecutar_clinica():
    while True:
        print("\n" + "="*30)
        print("   SISTEMA CLÍNICA DENTAL   ")
        print("="*30)
        print(f"1. Atenderse con Dr. {dentista1.nombre}")
        print(f"2. Atenderse con Dra. {dentista2.nombre}")
        print("3. Salir del programa")

        opcion_doc = input("\nSeleccione una opción (1-3): ")

        if opcion_doc == "3":
            print("Saliendo del sistema... ¡Que tenga un buen día!")
            print("Programa realizado por Jaquez Andres = 27")
            break

        if opcion_doc == "1":
            d = dentista1
        elif opcion_doc == "2":
            d = dentista2
        else:
            print("⚠️ Opción inválida, intenta de nuevo.")
            continue

        # Menú de acciones para el dentista seleccionado
        while True:
            print(f"\n--- Panel de Control: {d.nombre} ---")
            print("a) Examinar boca")
            print("b) Realizar limpieza")
            print("c) Colocar ortodoncia")
            print("d) Volver al menú principal")

            accion = input("¿Qué acción desea realizar?: ").lower()

            if accion == "a":
                print(f"\n>>> {d.examinar_boca()}")
            elif accion == "b":
                print(f"\n>>> {d.realizar_limpieza()}")
            elif accion == "c":
                print(f"\n>>> {d.colocar_ortodoncia()}")
            elif accion == "d":
                break
            else:
                print("⚠️ Acción no reconocida.")


if __name__ == "__main__":
    ejecutar_clinica()
```
